```python
import pandas as pd
import numpy as np
import os

PROJECT_ROOT = r'C:\Users\user\moscow_apartments'
df = pd.read_csv(os.path.join(PROJECT_ROOT, 'data', 'interim', 'apartments_cleaned.csv'))
print(df.shape)
df.head()
```

```python
df['floor_ratio'] = df['Floor']/df['Number of floors']

df['is_first_floor'] = (df['Floor'] == 1).astype(int)
df['is_last_floor'] = (df['Floor'] == df['Number of floors']).astype(int)

df['kitchen_area_ratio'] = df['Kitchen area']/df['Area']
```

```python
df['region_encoded'] = df['Region'].map({'Moscow region': 0,'Moscow':1})

df['apartment_type_encoded'] = df['Apartment type'].map({'Secondary':0,'New building':1})

df['renovation_encoded'] = df['Renovation'].map({'Without renovation': 0,'Cosmetic':1,'European-style renovation':2,'Designer':3})
```

```python
df['metro_station'] = df['Metro station'].astype('category')

cat_features = ['metro_station']
```

```python
df['log_price'] = np.log1p(df['Price'])
```

```python
os.makedirs(os.path.join(PROJECT_ROOT, 'data', 'processed'), exist_ok=True)
df.to_csv(os.path.join(PROJECT_ROOT, 'data', 'processed', 'apartments_features.csv'), index=False)
```
